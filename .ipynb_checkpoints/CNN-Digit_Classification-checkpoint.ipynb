# Numeric Digit Classification using Convolutional Neural Network

```python
import numpy as np
import cv2
import os
import random
import torch
import torch.nn as nn 
import torch.optim as optim
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
```

## 1. We define our device, whether we train our model on CPU or on GPU
If Nvidia-GPU found, model will be set to CUDA and train on our GPU, else it will be trained on our CPU.

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")
```

## 2. We now define our dataset and load it on basis of certain batch.
### Transforming our data to Grayscale, with 1 channel and converting it to tensor from a 2D image array.

```python
data_transforms = {
    'train' : transforms.Compose([transforms.Grayscale(1), transforms.ToTensor()]),
    'test'  : transforms.Compose([transforms.Grayscale(1), transforms.ToTensor()])
}

directory = 'Dataset/'
batch_size = 512
image_datasets = {x: datasets.ImageFolder(root=os.path.join(directory, x), transform = data_transforms[x]) for x in ['train', 'test']}
dataloaders = {x:torch.utils.data.DataLoader(image_datasets[x], batch_size=batch_size, shuffle=True) for x in ['train', 'test']}
```

## 3. Creating our Network. The Network architecture is made according to the following figure:

```python
class CNN(nn.Module):
    
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, stride=1)
        self.pool1 = nn.MaxPool2d(kernel_size=2)
        self.conv2 = nn.Conv2d(16, 32, 3, 1)
        self.relu = nn.ReLU()
        self.FC1 = nn.Linear(3872, 10)
        
        
    def forward(self, x):
        x = self.pool1(self.relu(self.conv1(x)))
        x = self.relu(self.conv2(x))
        x = x.view(x.size(0), -1) 
        x = self.FC1(x)
        return x
```

```python
net = CNN()
net = net.to(device)
```

## 4. We create a Classifier class with training function, loss function, a funtion to check accuracy, and a function to predict the output of a given input.

```python
class Digit_Classifier():

    def __init__(self, network, device, loss_func, image_datasets, dataloaders):
        self.net = network
        self.device = device
        self.image_datasets = image_datasets
        self.dataloaders = dataloaders  
        self.loss_func = loss_func
    
    def train(self, epochs, lr=0.01):
        optimizer = optim.SGD(net.parameters(), lr=lr, momentum=0.9)
        train_len = len(image_datasets['train'])
        test_len = len(image_datasets['test'])
        loss_arr = []
        for epoch in range(epochs):
            loss_main = 0.0
            train_acc = 0.0
            test_acc = 0.0
            for i, (images, labels) in enumerate(dataloaders['train']):
                output = self.predict(images, labels)
                optimizer.zero_grad()  
                acc = self.calc_accuracy(output, labels.to(device))
                train_acc += acc.item()
                loss = loss_func(output, labels.to(device))
                loss.backward()    
                loss_main += loss.item()
                optimizer.step()               

            for i, (image, labels) in enumerate(dataloaders['test']):
                output = self.predict(image, labels)      
                acc = self.calc_accuracy(output, labels.to(device))
                test_acc += acc
                
            print (f'Epoch {epoch+1}/{epochs} Loss: {loss_main} Train Accuracy: {train_acc/train_len} Test Accuracy: {test_acc/test_len}')
            loss_arr.append(loss_main)
        plt.plot(loss_arr)
    
        
    def calc_accuracy(self, y_pred, y_test):
        y_pred = torch.log_softmax(y_pred, dim=1)
        _, index = torch.max(y_pred, dim=1)
        true = (index == y_test).float()
        acc = true.sum()
        return acc

    def predict(self, image, labels):
        b_x = image.to(device)
        b_y = labels.to(device) 
        output = net(b_x)
        return output
    
    def predict_new(self, path):
        img = cv2.imread(path)
        img = transforms.ToTensor()(img)
        img = transforms.Grayscale()(img)
        img = img.unsqueeze(0)
        img = img.to(device)
        pred = self.net(img)
        pred = torch.argmax(pred)
        return pred
```

```python
loss_func = nn.CrossEntropyLoss()
classifier = Digit_Classifier(net, nn.CrossEntropyLoss(), device, image_datasets, dataloaders)
```

```python
classifier.train(epochs=10)
```

### The training accuracy can be seen of 97%+ and the testing accuracy is 97%+.

## 5. Now let's predict a few cases.

```python
ds_pred = {}

for i in range(10):
    path = f'Dataset/test/{i}'
    lists = os.listdir(path)
    ds_pred[i] = lists

image_of = int(input("Enter the digit that you want to read the image of: "))
index = np.random.randint(0, len(os.listdir(f'Dataset/test/{image_of}')))

im = np.array(ds_pred.get(image_of))
img_path = f'Dataset/test/{image_of}/{im[index]}'
res = classifier.predict_new(img_path)
print(f'Predicted: {res.item()} and Actual: {image_of}')
img = cv2.imread(img_path)
plt.matshow(img)
```
